```python
df1 = df1.drop("id",axis=1)
```

```python
df1
```

```python
input = df1.drop("diagnosis",axis=1).values
output = df1["diagnosis"].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(input,output,test_size=0.3,random_state=10)
```

```python
print("length of xtrain is:",len(x_train))
print("length of xtest is:",len(x_test))
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
print(x_train)
print()
print()
print(x_test)
```

```python
plt.figure(figsize=[10,10])
plt.plot(x_train)
plt.show()
```

```python
plt.figure(figsize=[10,10])
plt.plot(x_test)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
model1.fit(x_train,y_train)
```

```python
y_predict = model1.predict(x_test)
y_predict
```

```python
model1.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
a = confusion_matrix(y_predict,y_test)
a
```

```python
plt.figure(figsize=[10,10])
sn.heatmap(a,annot=True)
plt.show()
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(y_test,y_predict)
```

```python
print(classification_report(y_test,y_predict))
```
